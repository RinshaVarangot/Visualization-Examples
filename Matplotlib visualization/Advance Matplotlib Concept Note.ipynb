## Logarithmic Scale

A logarithmic scale is used when your data spans very large ranges (e.g., 1 to 1,000,000). Instead of equal spacing, it uses orders of magnitude (**powers of 10**).

In Matplotlib, you can apply a logarithmic scale to either the x-axis, the y-axis, or both. This is part of a broader transformation system that controls how data is displayed. To set the scale, you can use the **set_xscale()** and **set_yscale()** methods, and pass **"log"** as the argument.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,5,11)
y=x**2

fig,axes=plt.subplots(1,2,figsize=(10,4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic Scale(Y)")

plt.show()     
```

```python
## Another Example:

import matplotlib.pyplot as plt

a = [1, 2, 3, 4, 5]
b = [10, 100, 1000, 10000, 100000]

plt.plot(a, b)

plt.yscale('log')   # Apply log scale on Y-axis
plt.title("Logarithmic Scale Example")

plt.show()
```

## Scientific Notation

cientific notation is used to display very large or very small numbers in a compact form 

🔹 Why use it?

Makes numbers easier to read

Avoids long values like 1000000

Common in data science, finance, and scientific data

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker

formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```

## Axis number and axis label spacing

 distance between x and y axis and the numbers on the axes

```python
import matplotlib
matplotlib.rcParams['xtick.major.pad'] = 15
matplotlib.rcParams['ytick.major.pad'] = 15

fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("label and axis spacing")

# padding between axis label and axis numbers
ax.xaxis.labelpad = 15
ax.yaxis.labelpad = 15

ax.set_xlabel("x")
ax.set_ylabel("y");
```

```python
# restore defaults
matplotlib.rcParams['xtick.major.pad'] = 3
matplotlib.rcParams['ytick.major.pad'] = 3
```

## Axis position adjustments

Axis position adjustment means changing the location of the axes (top, bottom, left, right) or moving them inside/outside the plot for better visualization.

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.subplots_adjust(left=0.15, right=0.9, bottom=0.1, top=0.9);
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.subplots_adjust(left=0.25, right=.7, bottom=0.1, top=0.9);
```

## Axis grid

An axis grid adds horizontal and/or vertical lines across the plot, helping you read values more easily.

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='g', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
fig, ax = plt.subplots(figsize=(6,2))

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')

ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

# turn off axis spine to the right
ax.spines['right'].set_color("none")
ax.yaxis.tick_left() # only ticks on the left side
```

## Twin axes
Sometimes it is useful to have dual x or y axes in a figure; for example, when plotting curves with different units together. Matplotlib supports this with the twinx and twiny functions:

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

## Axes where x and y is zero

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```

## Other 2D plot styles
Apart from the regular plot() method, Matplotlib provides many other functions to create different types of 2D plots. Some of the most commonly used ones are listed below.

 See the matplotlib plot gallery for a complete list of available plot types: http://matplotlib.org/gallery.html. Some of the more useful ones are show below:



```python
n = np.array([0,1,2,3,4,5])
```

### Scatter Plot
A scatter plot is used to show the relationship between two variables by plotting points on a graph.

```python
fig, axes = plt.subplots( figsize=(12,3))

axes.scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes.set_title("scatter")
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.scatter(x, y, color='blue', s=100, alpha=0.7)

# color → point color
# s → size of points
# alpha → transparency
```

```python
plt.scatter(x, y, marker='x')

# Common markers:

# 'o' → circle
# 'x' → cross
# '^' → triangle
```

### Step Plot
A step plot is used to display data as discrete steps instead of continuous lines. It shows how values change at specific intervals.

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.step(n, n**2, lw=2)
axes.set_title("step")
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.step(x, y)
plt.title("Step Plot")
plt.show()
```

### Bar Plot
A bar plot is used to represent categorical data with rectangular bars, where the height (or length) shows the value.

```python
Fig,axes=plt.subplots(figsize=(12,3))

axes.bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes.set_title("bar")
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### Area Plot
An area plot is used to show trends over time, where the area under the line is filled with color

```python
fig,axes=plt.subplots(figsize=(12,3))

x = np.linspace(0, 5, 11)
y = x ** 2

axes.fill_between(x, x**2, x**3, color="green", alpha=0.5)
axes.set_title("fill_between");
```

```python
import matplotlib.pyplot as plt

a= [1,2,3,4,5]
b = [10,20,15,25,30]

plt.fill_between(a, b)
plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

## Text Annotation
Text annotation is used to add labels, notes, or highlights directly on a plot to explain important points.

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)

ax.text(0.15, 0.2, r"$y=x^2$", fontsize=20, color="blue")
ax.text(0.65, 0.1, r"$y=x^3$", fontsize=20, color="green");
```

## Figures with multiple subplots and insets
Axes can be added to a matplotlib Figure canvas manually using fig.add_axes or using a sub-figure layout manager such as subplots, subplot2grid, or gridspec

## subplots

```python
fig, ax = plt.subplots(2, 3)
fig.tight_layout()
```

## subplot2grid

```python
fig = plt.figure()

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))

fig.tight_layout()
```

```python
fig = plt.figure()

ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (0,2), rowspan=3)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))

fig.tight_layout()
```

## gridspec

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure()

gs = gridspec.GridSpec(2, 3, height_ratios=[2,1], width_ratios=[1,2,1])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
```

### Add Axes

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)
fig.tight_layout()

# inset
inset_ax = fig.add_axes([0.2, 0.55, 0.35, 0.35]) # X, Y, width, height

inset_ax.plot(xx, xx**2, xx, xx**3)
inset_ax.set_title('zoom near origin')

# set axis range
inset_ax.set_xlim(-.2, .2)
inset_ax.set_ylim(-.005, .01)

# set axis tick locations
inset_ax.set_yticks([0, 0.005, 0.01])
inset_ax.set_xticks([-0.1,0,.1]);
```
